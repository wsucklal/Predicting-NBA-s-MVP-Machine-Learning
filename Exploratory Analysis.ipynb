# Exploratory Analysis

```python
# Add a mean line later

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = 'master_table_final_fatigue.csv'
data = pd.read_csv(file_path)

# Plot MVP share
plt.figure(figsize=(10, 6))
sns.histplot(data['Share'], bins=30, kde=True)
plt.title('Distribution of MVP Share Values')
plt.xlabel('MVP Share')
plt.ylabel('Frequency')
plt.show()
```

```python
# Dropp uneeded/duplicate columns
columns_to_drop = ['Pts Won', 'First', 'Rank', 'Pts Max', 'WS', 'WS/48']
data_for_correlation = data.drop(columns=columns_to_drop)

# Correlation matrix
correlation_matrix = data_for_correlation.corr()
correlations_with_share = correlation_matrix['Share'].sort_values(ascending=False)
print(correlations_with_share)

plt.figure(figsize=(10, 8))
correlation_values = correlations_with_share.drop('Share')

soft_blue = '#1f77b4'
soft_red = '#d62728'

# Coloring bars based on their correlation value
colors = [soft_red if x < 0 else soft_blue for x in correlation_values.values]

correlation_values.plot(kind='bar', color=colors)
plt.title('Correlation of Features with MVP Share')
plt.xlabel('Features')
plt.ylabel('Correlation Coefficient')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
data_for_correlation = data.drop(columns=columns_to_drop)

def plot_with_best_fit(dataframe, feature_name, target='Share'):
    if pd.api.types.is_numeric_dtype(dataframe[feature_name]):
        plt.figure(figsize=(10, 6))
        sns.regplot(x=feature_name, y=target, data=dataframe, line_kws={"color":"red"})
        plt.title(f'{target} vs. {feature_name}')
        plt.xlabel(feature_name)
        plt.ylabel(target)
        plt.show()
    else:
        print(f"Skipping non-numeric feature: {feature_name}")

# Iterating over all features to create scatter plots
for feature in data_for_correlation.columns:
    if feature != 'Share':
        plot_with_best_fit(data_for_correlation, feature)
```

```python
featureName = 'Value Over Replacement Player'
feature='value_over_replacement_player'

mvp_data = data[data['Rank'] == 1]
other_data = data[data['Rank'] != 1]

# Plot other players in blue
sns.scatterplot(x=feature, y='Share', data=other_data, color=soft_blue, label='MVP Candidates')
    
# Plot MVP winners in red
sns.scatterplot(x=feature, y='Share', data=mvp_data, color='red', label='MVP Winners')

# Add a regression line for the entire dataset
sns.regplot(x=feature, y='Share', data=data, scatter=False, line_kws={"color": "black"})

plt.title(f'MVP Share vs. {featureName}')
plt.xlabel(featureName)
plt.ylabel('MVP Share')
plt.ylim(bottom=0)
plt.legend() 
plt.show()
```
